```python
# This is an example of Morgan fingerprints as implemented in RDKit
# For more information check:
# https://www.rdkit.org/docs/GettingStartedInPython.html
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import MolFromSmiles,MolToSmiles
from rdkit.Chem import Draw
from rdkit.Chem.Draw import rdMolDraw2D
# Glycidol
# mol=MolFromSmiles('OC[C@@H]1CO1')
# 2-benzyloxirane
smi='c1ccc(C[CH]2CO2)cc1'
mol=MolFromSmiles(smi)

# Prepare an interpretable Morgan Fingerprint instance
info={}
fp = AllChem.GetMorganFingerprint(mol,2,bitInfo=info)

# info will be a dictionary with bit id keys and 
# values of (atom index, radius)
print(info)
# The first bit, 98513984, is set three times: 
#    ((1, 1), (0, 1), (9, 1)).
# This means that it is set by atom 1, 0, and 9 at radius 1.
```

```python
env = Chem.FindAtomEnvironmentOfRadiusN(mol,1,3)
amap={}
patt=Chem.PathToSubmol(mol,env,atomMap=amap)

hit_ats = list(mol.GetSubstructMatch(patt))
hit_bonds = []
for bond in patt.GetBonds():
    aid1 = hit_ats[bond.GetBeginAtomIdx()]
    aid2 = hit_ats[bond.GetEndAtomIdx()]
    hit_bonds.append(mol.GetBondBetweenAtoms(aid1,aid2).GetIdx())
d = rdMolDraw2D.MolDraw2DCairo(500, 500) # or MolDraw2DCairo to get PNGs
d.drawOptions().addStereoAnnotation = True
d.drawOptions().addAtomIndices = True
# d.DrawMolecule(mol)
d.FinishDrawing()
# rdMolDraw2D.PrepareAndDrawMolecule(d, mol, highlightAtoms=[3])

rdMolDraw2D.PrepareAndDrawMolecule(d, mol, highlightAtoms=hit_ats,highlightBonds=hit_bonds)
d.WriteDrawingText('./atom_annotation_r1.png') 
```
