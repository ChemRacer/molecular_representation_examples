```python
import os
import sys
# !{sys.executable} -m pip install imageio
import numpy as np
from numpy.random import default_rng
rng = default_rng(42)  # Create a random number generator

from scipy.spatial.distance import pdist, squareform
from scipy.sparse import coo_matrix

from gtda.graphs import GraphGeodesicDistance
from gtda.homology import VietorisRipsPersistence, SparseRipsPersistence, FlagserPersistence

from igraph import Graph

from IPython.display import SVG, display
import networkx as nx
import matplotlib.pyplot as plt
import imageio
```

```python
# Create a single weighted adjacency matrix of a FCW graph
n_vertices = 5
proba_0 = 1/4

# x = rng.choice([0,1/2, 1,10], size=(n_vertices, n_vertices), p=[proba_0,proba_0, proba_0,proba_0])
# np.fill_diagonal(x, 0)
# x=np.maximum( x, x.transpose() )
x = np.array([[ 0, 10,  1,  1,  1],
       [10,  0,  3,  1, 10],
       [ 1,  3,  0, 10,  1],
       [ 1,  1, 10,  0,  1],
       [ 1, 10,  1,  1,  0]],dtype=float)
xog=x.copy()
x[x==0]=np.inf

# x = np.array([[0,2,0,4,2],
#        [2,0,1,3,0],
#        [0,1,0,1,0],
#        [4,3,1,0,2],
#        [2,0,0,2,0]],dtype=float)


# x = np.triu(x)



G = nx.from_numpy_array(xog)

# Fill the diagonal with zeros (not always necessary, see below)

import gudhi
f, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,6))
rips_complex = gudhi.RipsComplex(distance_matrix=x)
# simplex tree,
simplex_tree = rips_complex.create_simplex_tree(max_dimension=2)
# and persistence data from the point cloud
diag = simplex_tree.persistence(min_persistence=0)
# Generate the persistence barcode.
pb = gudhi.plot_persistence_barcode(diag, max_intervals=0,alpha=1.0,legend=True,axes=ax1,fontsize=12)
ax1.set_ylim(-1,5)
ax1.set_xlim(0,11)
# Generate the persistence diagram.
perd = gudhi.plot_persistence_diagram(diag, max_intervals=0,alpha=1.0,legend=True,axes=ax2,fontsize=12)
plt.tight_layout()
plt.savefig('persistG.png',dpi=300)
```

```python
pos=nx.spring_layout(G)
nx.draw_networkx(G,pos)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
plt.axis('off')
plt.tight_layout()
plt.savefig('G.png',dpi=300)
plt.show()
```

```python
K=nx.Graph()
filenames=[]
for k,v in {f1:[v for v,k in simplex_tree.get_skeleton(1) if k==f1 and k!=np.inf] for f1 in sorted(filters) if f1!=np.inf}.items():
    if k==0:
        edges = list(G.edges)
        G.remove_edges_from(edges)
        K=G.subgraph(val[0] for val in v if len(val)!=2)
    else:
        G.add_weighted_edges_from([(val[0],val[1],weight_dict[tuple(val)]) for val in v if len(val)==2])
        K=G
        for v2 in v:
            if len(v2)==3:
                tri=np.array([pos[z] for z in v2])
                trif=np.vstack([tri,tri[0]])
                trix=trif[:,0]
                triy=trif[:,1]
                plt.fill(trix,triy,'red',alpha=0.1)


    nx.draw_networkx(K,pos)
    labels = nx.get_edge_attributes(K,'weight')

    nx.draw_networkx_edge_labels(K,pos,edge_labels=labels)
    plt.axis('off')
    plt.title(f'Filter {k}')
    plt.tight_layout()
    filename=f'{k}_filterG.png'
    plt.savefig(filename,dpi=300)
    filenames.append(filename)
    plt.show()            
```

```python
with imageio.get_writer('graphgif.gif', mode='I',fps=1) as writer:
    for filename in filenames:
        image = imageio.v2.imread(filename)
        writer.append_data(image)

# Remove files
# for filename in set(filenames):
#     os.remove(filename)    
```
